```python
def load_classes(path):
    # Loads *.names file at 'path'
    with open(path, 'r') as f:
        names = f.read().split('\n')
    return list(filter(None, names))  # filter removes empty strings (such as last line)
```

```python
# yolor parameters

out = 'inference/output'
source = 'inference/images'
weights = 'yolor_data/best.pt'
view_img = False
save_img = False
save_txt = False
imgsz = 320
cfg = 'yolor_data/yolor_p6.cfg'
names = 'yolor_data/data.names'
webcam = source == '0' or source.startswith('rtsp') or source.startswith('http') or source.endswith('.txt')
device = ''
conf_thres = 0.4
iou_thres = 0.5
agnostic_nms = False
augment = False
update = False
classes = None

# Initialize
device = select_device(device)

# vit parameters
transformer = 'dino_vit_pretrained'
data_dir = 'cropped'
tire_multiplier = 2000
img_size = 224
```

```python
torch.cuda.empty_cache()

# Load vit model
vit = vits.__dict__['vit_small'](patch_size=16)
model_vit = vit.to(device)
model_vit.load_state_dict(torch.load('vit_weights/dino_vit_pretrained cropped 1-wheel blocked.pt'), strict = False)
model_vit.eval()
```

```python
# load MLP
from vit_pytorch_new.MLP import MLP
model_mlp = MLP(input_features=396, num_classes = 13).to(device)
model_mlp.load_state_dict(torch.load('mlp_weights/dino_vit_pretrained cropped 1-wheel blocked.pt'), strict = False)
model_mlp.eval()
```

```python
# Load yolor model
model_yolor = Darknet(cfg, imgsz).cuda()
model_yolor.load_state_dict(torch.load(weights, map_location=device)['model'])
model_yolor.to(device).eval()
half = device.type != 'cpu'  # half precision only supported on CUDA
if half:
    model_yolor.half()  # to FP16

# Set Dataloader
dataset = LoadImages(source, img_size=imgsz, auto_size=64)

# Get names and colors
names = load_classes(names)
colors = [[random.randint(0, 255) for _ in range(3)] for _ in range(len(names))]
```

```python
# Run inference
t0 = time.time()
# img = torch.zeros((1, 3, imgsz, imgsz), device=device)  # init img
for path, img, im0s, vid_cap in dataset:
    img = torch.from_numpy(img).to(device)
    img = img.half() if half else img.float()  # uint8 to fp16/32
    img_show = img.permute(1, 2, 0).detach().cpu().type(torch.int).numpy()
    img /= 255.0  # 0 - 255 to 0.0 - 1.0
    if img.ndimension() == 3:
        img = img.unsqueeze(0)
    with torch.no_grad():
        # Inference yolor
        t1 = time_synchronized()
        pred = model_yolor(img, augment=augment)[0]

        # Apply NMS
        pred = non_max_suppression(pred, conf_thres, iou_thres, classes=classes, agnostic=agnostic_nms)
        t2 = time_synchronized()

    # move class 1-13 to front, and remove confidence
    pred = pred[0].tolist()
    for bbox in reversed(pred):
        index = pred.index(bbox)
        pred[index].pop(4)
        for i in range(4):
            pred[index][i] = pred[index][i]/320    
        if pred[index][4] != 13:
            pred.insert(0, pred[index])
            pred.remove(pred[index+1])       
    
    # crop the image
    img_width = img.size(3)
    if len(pred) != 0:
        x1 = int(img_width*pred[0][0])
        y1 = int(img_width*pred[0][1])
        x2 = img_width*pred[0][2]
        y2 = img_width*pred[0][3]
        w = int(x2 - x1)
        h = int(y2 - y1)
    img = torchvision.transforms.functional.crop(img, y1, x1, h, w)
    img_resize = transforms.Resize((img_size, img_size))
    img = img_resize(img)
    # inference vit
    with torch.no_grad():
        features = model_vit(img.float())
    
    # get tire positions (asending)
    tires_pos = []
    if len(pred) >= 2:
        for i in range(1, len(pred)):
            if pred[i][4] == 13:
                x = (pred[i][2] + pred[i][0])/2
                tires_pos.append(x)
    tires_pos.sort()
    
    # relative positions
    for i in range(len(tires_pos)):
        if tires_pos[i] >= pred[0][0]:
            tires_pos[i] = tire_multiplier*(tires_pos[i]-pred[0][0])/(pred[0][2]-pred[0][0])
        else:
            tires_pos[i] = 0.0

    # concat vit features with tire positions
    total_features = torch.cat((features[0], torch.as_tensor(tires_pos).to(device), torch.zeros(12-len(tires_pos)).to(device)),0)
    predictions = model_mlp(total_features)
    pred_class = torch.argmax(predictions).item()+1
    t3 = time_synchronized()
    print('pred_class:',pred_class)
    print('time spent: %.3fs' % (t3 - t1))
print('Done. (%.3fs)' % (time.time() - t0))    
```
